## A. Use K-means to cluster each dataset

1. Create each copy of the dataset for k-means

```python
# put your answer here

df_kmeans_face = face_df.copy()
df_kmeans_chrome = chrome_df.copy()
df_kmeans_lines2 = lines2_df.copy()
df_kmeans_supernova = supernova_df.copy()
df_kmeans_network = network_df.copy()
df_kmeans_spirals = spirals_df.copy()
df_kmeans_densedisk = densedisk_df.copy()
df_kmeans_rings = rings_df.copy()
df_kmeans_chainlink = chainlink_df.copy()
```

2. Drop the "color" column on all datasets

```python
# put your answer here

df_kmeans_face.drop(['color'], axis=1, inplace=True)
df_kmeans_chrome.drop(['color'], axis=1, inplace=True)
df_kmeans_lines2.drop(['color'], axis=1, inplace=True)
df_kmeans_supernova.drop(['color'], axis=1, inplace=True)
df_kmeans_network.drop(['color'], axis=1, inplace=True)
df_kmeans_spirals.drop(['color'], axis=1, inplace=True)
df_kmeans_densedisk.drop(['color'], axis=1, inplace=True)
df_kmeans_rings.drop(['color'], axis=1, inplace=True)
df_kmeans_chainlink.drop(['color'], axis=1, inplace=True)
```

3. Train a K-means model for each dataset.

```python
# put your answer here

kmeans_face = KMeans(n_clusters=4, init='k-means++', random_state=42)
y_kmeans_face = kmeans_face.fit_predict(df_kmeans_face)

kmeans_chrome = KMeans(n_clusters=4, init='k-means++', random_state=42)
y_kmeans_chrome = kmeans_chrome.fit_predict(df_kmeans_chrome)

kmeans_lines2 = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans_lines2 = kmeans_lines2.fit_predict(df_kmeans_lines2)

kmeans_supernova = KMeans(n_clusters=4, init='k-means++', random_state=42)
y_kmeans_supernova = kmeans_supernova.fit_predict(df_kmeans_supernova)

kmeans_network = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans_network = kmeans_network.fit_predict(df_kmeans_network)

kmeans_spirals = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans_spirals = kmeans_spirals.fit_predict(df_kmeans_spirals)

kmeans_densedisk = KMeans(n_clusters=2, init='k-means++', random_state=42)
y_kmeans_densedisk = kmeans_densedisk.fit_predict(df_kmeans_densedisk)

kmeans_rings = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans_rings = kmeans_rings.fit_predict(df_kmeans_rings)

kmeans_chainlink = KMeans(n_clusters=2, init='k-means++', random_state=42)
y_kmeans_chainlink = kmeans_chainlink.fit_predict(df_kmeans_chainlink)
```

4. Add a "Cluster" column for each data set

```python
df_kmeans_face['Cluster'] = y_kmeans_face
df_kmeans_chrome['Cluster'] = y_kmeans_chrome
df_kmeans_lines2['Cluster'] = y_kmeans_lines2
df_kmeans_supernova['Cluster'] = y_kmeans_supernova
df_kmeans_network['Cluster'] = y_kmeans_network
df_kmeans_spirals['Cluster'] = y_kmeans_spirals
df_kmeans_densedisk['Cluster'] = y_kmeans_densedisk
df_kmeans_rings['Cluster'] = y_kmeans_rings
df_kmeans_chainlink['Cluster'] = y_kmeans_chainlink
```
